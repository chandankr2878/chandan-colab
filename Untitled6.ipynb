# rows columns selection

```python
# for column
import pandas as pd
df=pd.DataFrame({
    'name': ['chandan', 'vicky','arun'],
    "age": [23, 22,25],
    'salary':[10000,20000,30000]
})
print(df[['age']])
```

```python
#for row
import pandas as pd
df=pd.DataFrame({
    'name': ['chandan','vicky','vicky','arun'],
    "age": [23,24, 22,25],
    'salary':[10000,20000,30000]
})
print(df.loc[df.index==2])
```

```python
import pandas as pd
df=pd.DataFrame({
    'name': ['chandan','vicky','vicky','arun'],
    "age": [23,24, 22,25],
    'salary':[10000,15000,20000,30000]
})
print(df.loc[0:2])
print(df.iloc[0:2])
print(df.iloc[[0,1,2]])
```

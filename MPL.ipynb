```python
import matplotlib.pyplot as plt
x = [
    "apple", "banana", "orange", "mango",
    "grapes", "pineapple"
]

y = [10, 20, 30, 40, 50, 60]
# 1. Line Chart
plt.subplot(4,4,1)
plt.plot(x,y)
plt.title("Line")

# 2. Scatter Plot
plt.subplot(4,4,2)
plt.scatter(x,y)
plt.title("Scatter")

# 3. Bar Chart
plt.subplot(4,4,3)
plt.bar(x,y)
plt.title("Bar")

# 4. Horizontal Bar Chart
plt.subplot(4,4,4)
plt.barh(x,y)
plt.title("BarH")

# 5. Histogram
plt.subplot(4,4,5)
plt.hist(y)
plt.title("Histogram")

# 6. Pie Chart
plt.subplot(4,4,6)
plt.pie(y,labels=x)
plt.title("Pie")

# 7. Box Plot
plt.subplot(4,4,7)
plt.boxplot(y)
plt.title("Box")

# 8. Area Plot
plt.subplot(4,4,8)
plt.fill_between(x,y)
plt.title("Area")

# 9. Stack Plot
plt.subplot(4,4,9)
plt.stackplot(x,y)
plt.title("Stack")

# 10. Stem Plot
plt.subplot(4,4,10)
plt.stem(x,y)
plt.title("Stem")

# 11. Step Plot
plt.subplot(4,4,11)
plt.step(x,y)
plt.title("Step")

# 12. Error Bar Plot
plt.subplot(4,4,12)
plt.errorbar(x,y,yerr=2)
plt.title("Error")

# 13. Event Plot
plt.subplot(4,4,13)
plt.eventplot(y)
plt.title("Event")

# 14. Violin Plot
plt.subplot(4,4,14)
plt.violinplot(y)
plt.title("Violin")



# 16. Polar Plot
plt.subplot(4,4,15, projection='polar')
plt.plot(x,y)
plt.title("Polar")

plt.tight_layout()
plt.show()
```

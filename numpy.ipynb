```python
import numpy as np
# arrary 1d
arr = np.zeros(5)       #000000
print(arr)

# 2d
arr1 = np.zeros((2,3))+8 # 2 row and 3 cols
print(arr1)
```

```python

# ones
import numpy as np
# array 1d
arr = np.ones(5)
print(arr)

# 2d
arr1 = np.ones((2,3))+77
print(arr1)
```

```python
# full
import numpy as np
# array 1d
arr = np.full((5),7)
print(arr)

# 2d
arr1 = np.full((2,3),7+1)
print(arr1)
```

```python
import numpy as np
arr = np.zeros(5)
print(arr)
print(arr.shape) # 5
print(arr.ndim) # 1d
print(arr.size) # 5
print(arr.dtype) # float

# 2d
print("2d ones start")
arr1 = np.zeros((3,2))
print(arr1)
print(arr1.shape) # 3,2
print(arr1.ndim) # 2d
print(arr1.size) # 6
print(arr1.dtype) # float
```

```python
# 1d array arange
import numpy as np
arr = np.arange(6)
print(arr)

# 2d
arr1 = arr.reshape(2,3) # 2 rows and 3 cols
print(arr1)
```

```python
# flatten | copy
import numpy as np
# 2d
arr = np.arange(6)
arr1 = arr.reshape(2,3)
print(arr1)
# flatten
arr_new = arr1.flatten()
arr_new[0] = 100
print(arr1)
print(arr_new)
```

```python
# ravel | original
import numpy as np
# 2d
arr = np.arange(6).reshape(2,3)
print(arr)
# ravel
arr_new1 = arr.ravel()
arr_new1[0] = 200
print(arr)
print(arr_new1)
```

```python
# transpose
import numpy as np
arr = np.arange(9).reshape(3,3)
print(arr)

# transpose
arr1 = arr.transpose()
print(arr1)

#
arr1 = arr.T
print(arr1)
```

```python
# numpy array operations
# 1d
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr[2])
print(arr[0:4])#slicing
print(arr[-1])
print(arr[::])
print(arr[0:4:2])
print(arr[::2])# jump by 1 start from zero to end element
print(arr[::-1]) # for reverse

print(arr[:-3:-1])#########
```
